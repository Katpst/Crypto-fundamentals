### â­• [optional] Calculate and compare the execution time of different hash functions (MD5, SHA-1, SHA-256, SHA-512) on the same input.


````
import hashlib
import time

data = b"The quick brown fox jumps over the lazy dog"
iterations = 1_000_000

hash_functions = {
    "MD5": hashlib.md5,
    "SHA-1": hashlib.sha1,
    "SHA-256": hashlib.sha256,
    "SHA-512": hashlib.sha512
}

results = {}

for name, func in hash_functions.items():
    start = time.perf_counter()
    for _ in range(iterations):
        func(data).digest()
    end = time.perf_counter()
    results[name] = end - start

for name, t in results.items():
    print(f"{name:8s}: {t:.4f} seconds")
````

### â­•ï¸ Find the password that corresponds to the following hash
Help: the password contains less than 5 letters

````
import hashlib
import string
import itertools

target_hash = "071ebe8c4b384ff5bc3e35ef9037348584b3ef35a023282ea76862ec4edf213d"

def sha256(text):
    return hashlib.sha256(bytes(text, "UTF-8")).hexdigest()

def search_password(hash):
    chars = string.ascii_letters
    max_len = 5

    for length in range(1, max_len + 1):
        for candidate in itertools.product(chars, repeat=length):
            candidate = ''.join(candidate)
            if sha256(candidate) == hash:
                return candidate
    return None

print(search_password(target_hash))
````

ðŸŸ  How does the complexity of vanity increase with the target length?



The complexity of a vanity hash search increases **exponentially** with the length of the target.

A cryptographic hash (such as SHA-256) behaves like a random hexadecimal string. Each hexadecimal character has 16 possible values. Therefore, the probability that a hash starts with a specific target of length (k) is ((1/16)^k).

As a result, the expected number of attempts required to find a matching hash is (16^k). Each additional hexadecimal character in the target multiplies the required work by a factor of 16.

In terms of time complexity, a vanity search has complexity:

[
O(16^k)
]

This exponential growth makes short targets easy to find, while longer targets become computationally infeasible very quickly.

### â­•ï¸ [Optional] Implement a simplified proof-of-work system with difficulty adjustment based on the time it takes to mine a block


````
import hashlib
import time

def sha256_hex(s: str) -> str:
    return hashlib.sha256(s.encode("utf-8")).hexdigest()

def hash_to_int(h: str) -> int:
    return int(h, 16)

def block_header(prev_hash: str, data: str, timestamp: int, nonce: int) -> str:
    # Very simplified â€œheaderâ€
    return f"{prev_hash}|{data}|{timestamp}|{nonce}"

def mine_block(prev_hash: str, data: str, target: int):
    timestamp = int(time.time())
    nonce = 0
    start = time.perf_counter()

    while True:
        header = block_header(prev_hash, data, timestamp, nonce)
        h = sha256_hex(header)
        if hash_to_int(h) < target:   # proof-of-work condition
            elapsed = time.perf_counter() - start
            return {
                "prev_hash": prev_hash,
                "data": data,
                "timestamp": timestamp,
                "nonce": nonce,
                "hash": h,
                "elapsed": elapsed,
                "target": target,
            }
        nonce += 1

def adjust_target(target: int, elapsed: float, target_time: float,
                  min_factor: float = 0.25, max_factor: float = 4.0) -> int:
    """
    If mining was too fast (elapsed < target_time), make it harder -> smaller target.
    If mining was too slow (elapsed > target_time), make it easier -> larger target.
    """
    ratio = elapsed / target_time
    ratio = max(min_factor, min(max_factor, ratio))  # clamp changes

    new_target = int(target * ratio)

    # Keep target within valid range for a 256-bit hash
    max_target = (1 << 256) - 1
    new_target = max(1, min(max_target, new_target))
    return new_target

def is_valid_block(block: dict) -> bool:
    header = block_header(block["prev_hash"], block["data"], block["timestamp"], block["nonce"])
    h = sha256_hex(header)
    return (h == block["hash"]) and (hash_to_int(h) < block["target"])

# -----------------------
# Demo: mine N blocks with difficulty adjustment
# -----------------------
if __name__ == "__main__":
    N_BLOCKS = 6
    TARGET_TIME = 1.0  # seconds per block (adjust as you like)

    # Start fairly easy: target = 2**240 (smaller target = harder)
    target = 1 << 240

    chain = []
    prev = "0" * 64  # genesis previous hash

    for height in range(N_BLOCKS):
        data = f"Block {height} transactions"
        block = mine_block(prev, data, target)

        ok = is_valid_block(block)
        print(
            f"Height={height}  "
            f"time={block['elapsed']:.3f}s  "
            f"target=2^{block['target'].bit_length()-1}  "
            f"hash_prefix={block['hash'][:10]}...  "
            f"valid={ok}"
        )

        chain.append(block)
        prev = block["hash"]

        # difficulty adjustment for next block
        target = adjust_target(target, block["elapsed"], TARGET_TIME)

    print("Chain valid:", all(is_valid_block(b) for b in chain))
````

### ðŸŸ  Adapt the code above in order to be able to verify a signature from someone else


````
from ecdsa import VerifyingKey, SECP256k1, BadSignatureError

def verify_external_signature(public_key, signature_hex, message):
    try:
        vk = VerifyingKey.from_string(bytes.fromhex(public_key), curve=SECP256k1)
        signature = bytes.fromhex(signature_hex)
        vk.verify(signature, message.encode("utf-8"))
        return True
    except BadSignatureError:
        return False
````

### â­•ï¸ Calculate an Ethereum address from a private key


````
from ecdsa import SigningKey, SECP256k1
import hashlib

def eth_address_from_private_key(private_key):
    # private_key: hex string (32 bytes)
    sk = SigningKey.from_string(bytes.fromhex(private_key), curve=SECP256k1)

    # public key (64 bytes: x || y)
    vk = sk.get_verifying_key()
    public_key = vk.to_string()

    # SHA3-256 hash (used as Keccak-256 substitute)
    digest = hashlib.sha3_256(public_key).digest()

    # last 20 bytes
    return "0x" + digest[-20:].hex()
````

### â­•ï¸ Implement a simplified HD wallet that can derive a first keypair from a seed phrase.




````
import hashlib
import hmac
from ecdsa import SigningKey, SECP256k1

# Simplified HD-style wallet (NOT BIP-39/BIP-32 compatible; for exercises only)
# seed_phrase -> deterministic master seed -> child private key (index 0) -> keypair

def generate_wallet_from_seed(seed_phrase):
    # 1) Deterministic "seed" from the phrase
    seed = hashlib.sha256(seed_phrase.encode("utf-8")).digest()

    # 2) Derive first child key (index 0) using HMAC-SHA256
    index = 0
    child_material = hmac.new(
        key=seed,
        msg=index.to_bytes(4, "big"),
        digestmod=hashlib.sha256
    ).digest()

    # 3) Turn into a valid secp256k1 private key (1..n-1)
    n = SECP256k1.order
    priv_int = (int.from_bytes(child_material, "big") % (n - 1)) + 1
    priv_bytes = priv_int.to_bytes(32, "big")

    # 4) Build keypair
    sk = SigningKey.from_string(priv_bytes, curve=SECP256k1)
    vk = sk.get_verifying_key()

    return {
        "private_key_hex": priv_bytes.hex(),
        "public_key_hex": vk.to_string().hex(),  # 64 bytes (x||y)
        "index": index
    }

# Example
seed_phrase = "correct horse battery staple"
wallet = generate_wallet_from_seed(seed_phrase)
print(wallet)
````
